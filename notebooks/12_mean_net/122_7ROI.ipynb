```python
def reorder_corrs(corrmtx,labeldata,labels):
    """
    reorder correlation matrix according to network labels
    """

    idx=np.lexsort(([i for i in range(labeldata.shape[0])],labeldata[labels]))
    tmp=corrmtx[:,idx]
    return(tmp[idx,:],labeldata.iloc[idx,:])

def plot_reordered_corrs(corrmtx,labeldata,labels,colorbar=True):
    """
    plot correlation matrix after reordering
    """

    corr_reord,labeldata_reord=reorder_corrs(corrmtx,labeldata,labels)
    plt.imshow(corr_reord)
    # find breakpoints and plot lines
    breaks=np.array([int(not i) for i in labeldata_reord[labels].values[:-1]==labeldata_reord[labels].values[1:]])
    breaklocs=np.hstack((np.where(breaks)[0],np.array(corrmtx.shape[0]-1)))
    for b in breaklocs:
        plt.plot([0,corrmtx.shape[0]-1],[b,b],color='w',linewidth=0.5)
        plt.plot([b,b],[0,corrmtx.shape[0]-1],color='w',linewidth=0.5)
    # find label locations
    # add a zero to help find label locations
    breaklocs2=np.hstack(([0],breaklocs))
    label_locs=np.mean(np.vstack((breaklocs,breaklocs2[:-1])),0)
    networks=labeldata_reord[labels].values[breaklocs]
    ax=plt.gca()
    ax.set_yticks(label_locs)
    ax.set_yticklabels(networks, fontsize=20)
    if colorbar:
        plt.colorbar()
```

### Centerality measures of each module

```python
def strength(G):
    import networkx as nx
    return (dict(G.degree(weight= 'weight')))
```

```python
Mod_loc_mes = {}
Mod_glob_mes = {}
#for Mod in ['Visual']:
for Mod in nd:
    Mod_loc_mes[Mod] = {}
    Mod_glob_mes[Mod] = {}

    for g in subjects_groups:
        Mod_loc_mes[Mod][g] = {}
        Mod_glob_mes[Mod][g] = {}

        M = meanM_bin[g].copy()
        M = M[nd[Mod],:]
        M = M[:,nd[Mod]]
        G = nx.from_numpy_array(M)
        Mod_loc_mes[Mod][g]['stren'] = np.fromiter(strength(G).values(), dtype=float)
        Mod_loc_mes[Mod][g]['betw'] = np.fromiter(nx.betweenness_centrality(G).values(), dtype=float)
        Mod_loc_mes[Mod][g]['clus'] = np.fromiter(nx.clustering(G).values(), dtype=float)
        
        Mod_glob_mes[Mod][g]['average_clustering'] = nx.average_clustering(G)
        Mod_glob_mes[Mod][g]['local_efficiency'] = nx.local_efficiency(G)
        Mod_glob_mes[Mod][g]['global_efficiency'] = nx.global_efficiency(G)
        Mod_glob_mes[Mod][g]['pearson_correlation'] = nx.degree_pearson_correlation_coefficient(G)
```

```python
from matplotlib import rc
def bar_plot(df,bar_names,width,height,fontsize):

    totals = [i+j+k+l for i,j,k,l in zip(df['AD'], df['LMCI'], df['EMCI'], df['CN'])]
    ADp = [i / j * 100 for i,j in zip(df['AD'], totals)]
    LMCIp = [i / j * 100 for i,j in zip(df['LMCI'], totals)]
    EMCIp = [i / j * 100 for i,j in zip(df['EMCI'], totals)]
    CNp = [i / j * 100 for i,j in zip(df['CN'], totals)]
    
    n_bars = df.shape[0]
    ind = np.arange(n_bars)
    plt.figure(figsize=(height,width))
    ax = plt.axes(frameon=False)

    # plot
    barWidth = 0.85
    # Create green Bars
    plt.bar(ind, ADp, color='#e7305b', edgecolor='white', width=barWidth,label='AD')
    # Create orange Bars
    plt.bar(ind, LMCIp, bottom=ADp, color='#e2979c', edgecolor='white', width=barWidth,label='LMCI')
    # Create blue Bars
    plt.bar(ind, EMCIp, bottom=[i+j for i,j in zip(LMCIp, ADp)], color='#e1d97e', edgecolor='white', width=barWidth,label='EMCI')
    plt.bar(ind, CNp, bottom=[i+j+k for i,j,k in zip(EMCIp, LMCIp, ADp)], color='#9bdeac', edgecolor='white', width=barWidth,label='CN')

    for xpos, ypos, yval in zip(ind, np.array(ADp)/2, df['AD'].values):
        plt.text(xpos, ypos, 
                 'AD \n' + str(np.around(yval,3)),
                 ha="center", va="center",fontsize=fontsize)
    for xpos, ypos, yval in zip(ind, np.array(ADp)+np.array(LMCIp)/2, df['LMCI'].values):
        plt.text(xpos, ypos,
                 'LMCI \n' + str(np.around(yval,3)),
                 ha="center", va="center",fontsize=fontsize)
    for xpos, ypos, yval in zip(ind, np.array(ADp)+np.array(LMCIp)+np.array(EMCIp)/2, df['EMCI'].values):
        plt.text(xpos, ypos,
                 'EMCI \n' + str(np.around(yval,3)),
                 ha="center", va="center",fontsize=fontsize)
    for xpos, ypos, yval in zip(ind, np.array(ADp)+np.array(LMCIp)+np.array(EMCIp)+np.array(CNp)/2, df['CN'].values):
        plt.text(xpos, ypos,
                 'CN \n' + str(np.around(yval,3)),
                 ha="center", va="center",fontsize=fontsize)

    # Custom x axis
    plt.xticks(ind, bar_names)
    ax.get_xaxis().tick_bottom()
    ax.axes.get_yaxis().set_visible(False)

    plt.legend(subjects_groups[::-1],loc='upper left', bbox_to_anchor=(1,1), ncol=1)
    plt.show()
    return
```
