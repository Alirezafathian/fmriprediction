```python
groupcolor_def = [(144/255,29/255,158/255),(73/255,142/255,196/255),(0/255,122/255,13/255),(216/255,75/255,243/255),(144/255,151/255,113/255),(205/255,130/255,33/255),(213/255,71/255,88/255),(1,1,1)]
from config import *
from src.data.materials import *
from src.data import subjects
import numpy as np
import nibabel as nb
import pandas as pd
import glob,os
import random
import scipy.stats
from nilearn import surface,datasets,plotting
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
def brain_viz_groupnodes(nodes,title,method,node_size,groupcolor= [],marker = "o"):
    """
    """
    global coordinates
    groups = list(nodes.keys())
    NEWcoordinates = np.vstack([coordinates[nodes[g],:] for g in groups])
    NEWcoordinates = np.vstack([NEWcoordinates,coordinates[3,:]])
    node_size1 = np.hstack([node_size[i] for i in range(len(node_size))])
    node_size1 = np.hstack([node_size1,np.array([0])])
    empty_index = len(node_size1)-1
    node_size1 = node_size1.astype('float32')
    node_size1-=np.mean(node_size1)
    if np.std(node_size1)!=0:
        node_size1/=np.std(node_size1)
    node_color = []
    if groupcolor == []:
        groupcolor = groupcolor_def
    for i in range(len(groups)):
        a = [groupcolor[i]] * len(nodes[groups[i]])
        node_color.extend(a)
    node_color.extend([(1,1,1)])

    NEWnodes = np.hstack([nodes[g] for g in groups])
    c0 = []
    for i in np.unique(NEWnodes):
        if len(np.where(NEWnodes==i)[0])>1:
            loc = np.where(NEWnodes==i)[0]
            s0 = node_size1[loc]
            a = np.argsort(s0)[::-1]
            s0 = s0[a]
            c0 = [node_color[i] for i in loc[a]]
            coor0 = np.full((len(np.where(NEWnodes==i)[0]),3), coordinates[i,:])
        if c0!=[]:
            NEWcoordinates = np.vstack([NEWcoordinates,coor0])
            node_size1 = np.hstack([node_size1,s0])
            node_color.extend(c0)

    if method == '2d':
        print(title)
        #node_size1=np.sqrt((node_size1 * 15)**5)
        node_size1=(node_size1 ** 8)* 250
        node_size1+=1200.0
        node_size1[empty_index] = 0.0
        node_kwargs = {}
        node_kwargs["marker"] = marker
        #node_kwargs["edgecolor"] = 'black'
        node_kwargs["linewidth"] = 4
        view=plotting.plot_connectome(np.zeros((len(NEWcoordinates),len(NEWcoordinates))),
            NEWcoordinates,edge_threshold=0, node_size=node_size1,
            node_color = node_color,axes = (0, 0, 3, 3),node_kwargs = node_kwargs)
        view
    elif method == '3d':
        node_size1=node_size1*5.0
        node_size1+=10
        node_size1[empty_index] = 0.0
        view=plotting.view_markers(NEWcoordinates,title = title ,marker_size=node_size1,marker_color=node_color)
        view.open_in_browser()
        view
    return node_size1
```

```python
def smlwrld(G,Grand):
    Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
    G0 = G.subgraph(Gcc[0])
    C = nx.average_clustering(G0, nodes=None, weight=True, count_zeros=True)
    L = nx.average_shortest_path_length(G0, weight=True)
    Lr= nx.average_shortest_path_length(Grand, weight=True)
    Cr= nx.average_clustering(Grand, nodes=None, weight=True, count_zeros=True)
    return (C / Cr) / (L / Lr)
```
